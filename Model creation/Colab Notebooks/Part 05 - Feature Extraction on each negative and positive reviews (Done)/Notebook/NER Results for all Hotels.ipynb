```python
import os

# Function to create the anchor.txt file if it doesn't exist
def create_anchor_file(input_data_file):
    anchor_file_path = os.path.join(input_data_file)

    if not os.path.isfile(anchor_file_path):
        with open(anchor_file_path, 'w') as file:
            file.write('0')


def create_time_file(input_data_file):
    anchor_file_path = os.path.join(input_data_file)

    if not os.path.isfile(anchor_file_path):
        with open(anchor_file_path, 'w') as file:
            file.write('0')

# Function to reset the anchor.txt content to 0
def reset_anchor_file(reset_all=False):
    if reset_all:
        with open(InputDatalocation , 'w') as file:
            file.write('0')
        with open(InputDatalocationTotalTime , 'w') as file:
            file.write('0')

# Function to read the anchor.txt content, increment by 1, and return the updated value
def increment_anchor_file_value(final_df_instance):
    if final_df_instance is None:
        return

    with open(InputDatalocation , 'r+') as file:
        content = file.read()
        value = int(content)
        value += 1
        file.seek(0)
        file.write(str(value))
        return value


# Function to read the anchor.txt content, increment by 1, and return the updated value
def increment_time_file_value(time):
    if time is None:
        return

    time_seconds = int(time)  # Convert time to seconds

    with open(InputDatalocationTotalTime, 'r+') as file:
        content = file.read()

        if content.strip() == '0':
            value = time_seconds
        else:
            value = int(content) + time_seconds

        file.seek(0)
        file.write(str(value))
        return value

def read_time_file_value():
    with open(InputDatalocationTotalTime, 'r') as file:
        content = file.read().strip()
        if content == '0':
            return 0
        else:
            return int(content)
```

```python
def load_strings_from_text_file(text_file_path):
    with open(text_file_path, 'r') as file:
        lines = file.readlines()
        strings_list = [line.strip() for line in lines]
    return strings_list

def combine_and_save_strings_to_text_file(text_file_path, new_strings_list):
    existing_strings_list = load_strings_from_text_file(text_file_path)
    combined_set = set(existing_strings_list + new_strings_list)
    combined_list = list(combined_set)

    with open(text_file_path, 'w') as file:
        for string in combined_list:
            file.write(string + '\n')
```

**Food NER code**

```python
def get_food_NER(example):
  # Get food entities using the NER model
  ner_entity_results = pipe(example)

  # Create a list to store filtered food entities from the NER results
  filtered_food_entities_ner = []
  for result in ner_entity_results:
      entity = result["word"]
      if entity.lower() in foods_array:
          filtered_food_entities_ner.append(entity.lower())

  # Function to extract food entities from text using the foods_array
  def extract_food_entities(text):
      tokens = word_tokenize(text.lower())
      food_entities = []
      for food in foods_array:
          food_tokens = word_tokenize(food)
          if all(token in tokens for token in food_tokens):
              food_entities.append(food)
      return food_entities

  # Get food entities using nltk approach
  filtered_food_entities_nltk = extract_food_entities(example)

  # Combine both results to get the final list of food entities
  final_food_entities = list(set(filtered_food_entities_ner + filtered_food_entities_nltk))

  # print("\nFinal list of food entities:")
  # print(final_food_entities)
  # print(len(final_food_entities))

  return final_food_entities
```

```python
def organize_entities(data):
    result_dict = {}

    for entry in data:
        entity_group = entry['entity_group']
        word = entry['word']

        if entity_group in result_dict:
            result_dict[entity_group].append(word)
        else:
            result_dict[entity_group] = [word]

    return result_dict



def add_foods(entities_dict, food_list):
    if "FOODS" in entities_dict:
        entities_dict["FOODS"].extend(food_list)
    else:
        entities_dict["FOODS"] = food_list
```

```python
import time

def calculate_time_taken(start_time, end_time):
    # Calculate the time difference in seconds
    time_diff = end_time - start_time

    # Convert the time difference to a human-readable format
    hours = int(time_diff / 3600)
    minutes = int((time_diff % 3600) / 60)
    seconds = int(time_diff % 60)

    # Return the formatted time difference
    return f"{hours} hours, {minutes} minutes, {seconds} seconds"
```
