**Importing trequired libraries**

```python
import csv
import pandas as pd
import numpy as np
```

**Loading data.csvs to pandas dataframes**

```python
# Read the CSV file using pandas
Positivedata = pd.read_csv(InputPositiveDataFile)
Negativedata = pd.read_csv(InputNegativeDataFile)
PolaritiesAddeddata = pd.read_csv(PolaritiesAddedDataset)
# number_of_training_cases = len(data.index)
print(len(Positivedata.index))
print(len(Negativedata.index))
print(len(PolaritiesAddeddata.index))
```

**Adding monotoncally increasing sequance number**

```python
# Adding a monotonically increasing sequence number as a new column
PolaritiesAddeddata['Sequence'] = [i for i, _ in enumerate(PolaritiesAddeddata.iterrows(), start=1)]

# Displaying the modified dataframe
display(PolaritiesAddeddata.iloc[1:3])
```

**Getting only the required clomuns from each dataframe**

```python
import pandas as pd

# Indexes of the columns you want to select (0-based index)
selected_column_indexes = [1, 0, 3 ,5 ]

# Header topics for the new DataFrame
Positiveheader_topics = ['Sequence', 'Uncleaned Positive summmary', 'Original Positive review', 'Cleaned Positive Summmary']
Negativeheader_topics = ['Sequence', 'Uncleaned Negative summmary', 'Original Negative review', 'Cleaned Negative Summmary']

# Create a new DataFrame by selecting columns by their index
Positivedata_df_selected = Positivedata.iloc[:, selected_column_indexes].copy()
Negativedata_df_selected = Negativedata.iloc[:, selected_column_indexes].copy()

# Assign header topics to the new DataFrame columns
Positivedata_df_selected.columns = Positiveheader_topics
Negativedata_df_selected.columns = Negativeheader_topics

```

**Column headers are printed**

```python
# Getting required column names
column_headers = Positivedata_df_selected.columns
# Output the column headers
print(column_headers)

# Getting required column names
column_headers = Negativedata_df_selected.columns
# Output the column headers
print(column_headers)

# Getting required column names
column_headers = PolaritiesAddeddata.columns
# Output the column headers
print(column_headers)
```

**All are joined based on Sequence nummber**

```python
import pandas as pd

# Assuming you have already loaded your data into the 'PolaritiesAddeddata' and 'Positivedata_df_selected' dataframes

# Convert 'Sequence' column in Positivedata_df_selected to int64
# Positivedata_df_selected['Sequence'] = Positivedata_df_selected['Sequence'].astype('int64')

# Now you can perform the merge on the int64 'Sequence' columns
Positivedatamerged = pd.merge(PolaritiesAddeddata, Positivedata_df_selected, on='Sequence', how='inner')
FullyMerged = pd.merge(Positivedatamerged, Negativedata_df_selected, on='Sequence', how='inner')
```

```python
display(FullyMerged.iloc[300:302])
```

```python
# Getting required column names
column_headers = FullyMerged.columns
# Output the column headers
print(column_headers)
```

**Selecting only the required olumns oout of joined dataframe**

```python
# List of column names you want to select
columns_to_select = [
    'Sequence',
    'Hotel_Name_values',
    'Hotel_Address_values',
    'Review_Date_values',
    'Average_Score_values',
    'Reviewer_Nationality_values',
    'days_since_review',

    'Negative_Review',
    'Review_Total_Negative_Word_Counts',
    'negative_sentence_column_Sentiment',
    'negative_sentence_column_Positivity Percentage',
    'negative_sentence_column_Negativity Percentage',
    'Uncleaned Negative summmary',
    'Cleaned Negative Summmary',

    'Positive_Review',
    'Review_Total_Positive_Word_Counts',
    'positive_sentence_column_Sentiment',
    'positive_sentence_column_Positivity Percentage',
    'positive_sentence_column_Negativity Percentage',
    'Uncleaned Positive summmary',
    'Cleaned Positive Summmary',

    'Reviewer_Score',

    'latitude',
    'longitude'
    ]

# Create the new dataframe with selected columns
new_df = FullyMerged[columns_to_select].copy()
```

**Renaming them based on the requirement**

```python
# Dictionary to map old column names to new column names
column_name_mapping = {
    'Sequence'                                      :'Sequence',
    'Hotel_Name_values'                             :'Hotel_Name_values',
    'Hotel_Address_values'                          :'Hotel_Address_values',
    'Review_Date_values'                            :'Review_Date_values',
    'Average_Score_values'                          :'Average_Score_values',
    'Reviewer_Nationality_values'                   :'Reviewer_Nationality_values',
    'days_since_review'                             :'days_since_review',

    'Negative_Review'                               :'Negative_Review',
    'Review_Total_Negative_Word_Counts'             :'Review_Total_Negative_Word_Counts',
    'negative_sentence_column_Sentiment'            :'Negative_sentence_column_Sentiment',
    'negative_sentence_column_Positivity Percentage':'Negative_sentence_column_Positivity_Percentage',
    'negative_sentence_column_Negativity Percentage':'Negative_sentence_column_Negativity_Percentage',
    'Uncleaned Negative summmary'                   :'Uncleaned_Negative_summmary',
    'Cleaned Negative Summmary'                     :'Cleaned_Negative_Summmary',

    'Positive_Review'                               :'Positive_Review',
    'Review_Total_Positive_Word_Counts'             :'Review_Total_Positive_Word_Counts',
    'positive_sentence_column_Sentiment'            :'Positive_sentence_column_Sentiment',
    'positive_sentence_column_Positivity Percentage':'Positive_sentence_column_Positivity Percentage',
    'positive_sentence_column_Negativity Percentage':'Positive_sentence_column_Negativity Percentage',
    'Uncleaned Positive summmary'                   :'Uncleaned_Positive_summmary',
    'Cleaned Positive Summmary'                     :'Cleaned_Positive_Summmary',

    'Reviewer_Score'                                :'Reviewer_Score',

    'latitude'                                      :'latitude',
    'longitude'                                     : 'longitude'
}

# Renaming the selected columns
new_df.rename(columns=column_name_mapping, inplace=True)
```
