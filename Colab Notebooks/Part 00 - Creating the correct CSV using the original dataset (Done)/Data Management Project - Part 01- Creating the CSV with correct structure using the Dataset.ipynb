**Address to Dataset**

**Creating the initial dataframe to work with**

```python
# Open the file using the csv module
with open(InputDataFIleAddress, newline='', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')

    # Initialize empty dictionary to store the each column
    Dictionary_of_arrays = {
        "Hotel_Address_values": [],
        "Additional_Number_of_Scoring_values": [],
        "Review_Date_values": [],
        "Average_Score_values": [],
        "Hotel_Name_values": [],
        "Reviewer_Nationality_values": [],
        "Negative_Review": [],
        "Review_Total_Negative_Word_Counts": [],
        "Total_Number_of_Reviews": [],
        "Positive_Review": [],
        "Review_Total_Positive_Word_Counts": [],
        "Total_Number_of_Reviews_Reviewer_Has_Given": [],
        "Reviewer_Score": [],
        "tag_column": [],
        "days_since_review": [],
        "latitude": [],
        "longitude": []
    }

    # Loop through each row of the file and append each value to the corresponding list
    for row in csvreader:
        # Empty array to add tags
        tags_array=[]

        Dictionary_of_arrays["Hotel_Address_values"].append(row[0])
        Dictionary_of_arrays["Additional_Number_of_Scoring_values"].append(row[1])
        Dictionary_of_arrays["Review_Date_values"].append(row[2])
        Dictionary_of_arrays["Average_Score_values"].append(row[3])
        Dictionary_of_arrays["Hotel_Name_values"].append(row[4])
        Dictionary_of_arrays["Reviewer_Nationality_values"].append(row[5])
        Dictionary_of_arrays["Negative_Review"].append(row[6])
        Dictionary_of_arrays["Review_Total_Negative_Word_Counts"].append(row[7])
        Dictionary_of_arrays["Total_Number_of_Reviews"].append(row[8])
        Dictionary_of_arrays["Positive_Review"].append(row[9])
        Dictionary_of_arrays["Review_Total_Positive_Word_Counts"].append(row[10])
        Dictionary_of_arrays["Total_Number_of_Reviews_Reviewer_Has_Given"].append(row[11])
        Dictionary_of_arrays["Reviewer_Score"].append(row[12])
        Dictionary_of_arrays["days_since_review"].append(row[-3])
        Dictionary_of_arrays["latitude"].append(row[-2])
        Dictionary_of_arrays["longitude"].append(row[-1])
        # Loop through the "tags" values and apppend each into a single array
        for j in range(12,len(row)-3):
          tags_array.append(row[j])
        Dictionary_of_arrays["tag_column"].append(tags_array)

# Convert the list into a pandas DataFrame
df = pd.DataFrame({
    'Hotel_Address_values': Dictionary_of_arrays["Hotel_Address_values"],
    'Additional_Number_of_Scoring_values': Dictionary_of_arrays["Additional_Number_of_Scoring_values"],
    'Review_Date_values': Dictionary_of_arrays["Review_Date_values"],
    'Average_Score_values': Dictionary_of_arrays["Average_Score_values"],
    'Hotel_Name_values': Dictionary_of_arrays["Hotel_Name_values"],
    'Reviewer_Nationality_values': Dictionary_of_arrays["Reviewer_Nationality_values"],
    'Negative_Review': Dictionary_of_arrays["Negative_Review"],
    'Review_Total_Negative_Word_Counts': Dictionary_of_arrays["Review_Total_Negative_Word_Counts"],
    'Total_Number_of_Reviews': Dictionary_of_arrays["Total_Number_of_Reviews"],
    'Positive_Review': Dictionary_of_arrays["Positive_Review"],
    'Review_Total_Positive_Word_Counts': Dictionary_of_arrays["Review_Total_Positive_Word_Counts"],
    'Total_Number_of_Reviews_Reviewer_Has_Given': Dictionary_of_arrays["Total_Number_of_Reviews_Reviewer_Has_Given"],
    'Reviewer_Score': Dictionary_of_arrays["Reviewer_Score"],
    'tag_column': Dictionary_of_arrays["tag_column"],
    'days_since_review': Dictionary_of_arrays["days_since_review"],
    'latitude': Dictionary_of_arrays["latitude"],
    'longitude': Dictionary_of_arrays["longitude"]
})
```
